Computational tree

Solving the cable equation on a morpholgy is often expensive, hence, MorphTree defines a 'computational tree' that replaces the individual nodes between bifurcations with equivalent cylinders. A second tree of lower resolution is thus defined

```python
m_tree.set_comp_tree()
```

```python
for node in m_tree: print(str(node) + ', R (um) = ' + str(node.R) + 'L (um) = ' + str(node.L))
```

```python
with m_tree.as_computational_tree:
    for node in m_tree: print(str(node) + ', R (um) = ' + str(node.R) + 'L (um) = ' + str(node.L))
```

Locations 
locations on a morphology are defined relative to the original node structure. A location consists of two coordinates: the node index and an x-coordinate (x between 0 and 1). Specifying the relative position of the location between the node and its parent node (x= 0 means that the location is at the parent node's 3d coordinates whereas x=1 means that it's at the node's coordinates). Locations are instatiated from a tuple or dictionary

```python
loc1 = MorphLoc({'node': 4, 'x': .2}, m_tree)
loc2 = MorphLoc({5, .8}, m_tree)
```

```python
loc1_ = {'node': 4, 'x': .2}
loc2_ = {5, .8}
m_tree.store_locs([loc1_, loc2_], name='my favourite locs')
```
