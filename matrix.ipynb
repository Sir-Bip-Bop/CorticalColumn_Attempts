```python
import scipy.io
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import glob 
import os
```

```python
mats = {}
i = 0
for file in glob.glob(os.path.join('DTI_data/ACA/', '*.count.end.connectivity.mat')):
    mats[i] = scipy.io.loadmat(file)
    i += 1

mean_mat_connectivity = np.mean([mats[i]['connectivity'] for i in range(len(mats))], axis=0)

mats = {}
i = 0
for file in glob.glob(os.path.join('DTI_data/ACA/', '*.dti_fa.end.connectivity.mat')):
    mats[i] = scipy.io.loadmat(file)
    i += 1

mean_mat_dti = np.mean([mats[i]['connectivity'] for i in range(len(mats))], axis=0)

mats = {}
i = 0
for file in glob.glob(os.path.join('DTI_data/ACA/', '*.mean_length.end.connectivity.mat')):
    mats[i] = scipy.io.loadmat(file)
    i += 1

mean_mat_length = np.mean([mats[i]['connectivity'] for i in range(len(mats))], axis=0)

mats = {}
i = 0
for file in glob.glob(os.path.join('DTI_data/ACA/', '*.qa.end.connectivity.mat')):
    mats[i] = scipy.io.loadmat(file)
    i += 1

mean_mat_qa = np.mean([mats[i]['connectivity'] for i in range(len(mats))], axis=0)

fig, ax = plt.subplots(figsize=(17,14))

im = ax.imshow(mean_mat_connectivity,interpolation='none')
fig.colorbar(im,label='Number of tracts')
ax.set_title('Matriz de count end media sujetos ACA',fontsize=20)

fig, ax = plt.subplots(figsize=(17,14))

im = ax.imshow(mean_mat_dti,interpolation='none')
fig.colorbar(im,label='dti_fa')
ax.set_title('Matriz de fa media sujetos ACA',fontsize=20)

fig, ax = plt.subplots(figsize=(17,14))

im = ax.imshow(mean_mat_qa,interpolation='none')
fig.colorbar(im,label='qa')
ax.set_title('Matriz de qa media sujetos ACA',fontsize=20)

fig, ax = plt.subplots(figsize=(17,14))
im = ax.imshow(mean_mat_length,interpolation='none')
fig.colorbar(im,label='mean_length')
ax.set_title('Matriz de longitud media sujetos ACA',fontsize=20)
```
